```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df.drop_duplicates(inplace=True)

# Drop unnecessary columns (ignore if not found)
df.drop(['house_rules', 'license'], axis=1, inplace=True, errors='ignore')

# Clean currency columns
df['price'] = df['price'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False)
df['service fee'] = df['service fee'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False)

# Convert datatypes
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['service fee'] = pd.to_numeric(df['service fee'], errors='coerce')
df['id'] = df['id'].astype(str)
df['host id'] = df['host id'].astype(str)
df['last review'] = pd.to_datetime(df['last review'], errors='coerce')
df['Construction year'] = pd.to_numeric(df['Construction year'], errors='coerce').astype('Int64')

# Drop rows with missing critical values
df.dropna(subset=['price', 'service fee'], inplace=True)

# Keep only realistic availability values
df = df[df['availability 365'] <= 365]
```

```python
property_type=df['room type'].value_counts().to_frame()
property_type
```

```python
room_type_bar=plt.bar(property_type.index,property_type.loc[:,"count"]);
plt.bar_label(room_type_bar,labels=property_type.loc[:,"count"],padding=4);
plt.ylim([0,50000]);
plt.xlabel('Room Type');
plt.ylabel('Room type Count');
plt.title('Property Types and their count in the Dataset');
```

```python
import matplotlib.pyplot as plt
hood_group=df['neighbourhood group'].value_counts().to_frame()
hood_group
```

```python
hood_group_bar=plt.bar(hood_group.index,hood_group.loc[:,"count"]);
plt.bar_label(hood_group_bar,labels=hood_group.loc[:,"count"],padding=4);
plt.ylim([0,40000]);
plt.xlabel('Neighborhood Group');
plt.ylabel('Number of listings');
plt.xticks(rotation=45);
plt.title('which Neighborhood Groups has the highest no of listings');
```

```python
avg_price = df.groupby('neighbourhood group')['price'].mean().sort_values(ascending=False).to_frame()
avg_price_bar = plt.bar(avg_price.index, avg_price['price'])
plt.bar_label(avg_price_bar, labels=round(avg_price['price'], 2), label_type='edge', padding=4)
plt.ylim([0, 700])
plt.xlabel('Neighborhood Group')
plt.ylabel('Average Price ($)')
plt.title('Average Price by Neighborhood Group')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df.groupby('Construction year')['price'].mean().to_frame().plot()
plt.xlabel('Construction year')
plt.ylabel('Average Price ($)')
plt.title('Average Price ($) for properties in each construction year')
```

```python
hosts = (
    df.groupby('host name')['calculated host listings count']
    .sum()
    .sort_values(ascending=False)
    .nlargest(10)
    .to_frame()
)

hosts_bar = plt.bar(hosts.index, hosts['calculated host listings count'])
plt.bar_label(hosts_bar, labels=hosts['calculated host listings count'], label_type='edge', padding=3)
plt.xlabel('Host Name')
plt.ylabel('Calculated Host Listings Count')
plt.title('Top 10 Hosts by Number of Listings')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
review=df.groupby('host_identity_verified')['review rate number'].mean().sort_values(ascending=False).to_frame()
review
```

```python
review_bar=plt.bar(review.index,review.loc[:,"review rate number"]);
plt.bar_label(review_bar,labels=round(review.loc[:,"review rate number"],2),padding=4);
plt.ylim([0,4]);
plt.xlabel('Host Verification Status');
plt.ylabel('Average Review Rate Number');
plt.title('Average Review Rate for each Verification Statistics.');
```

```python
base_color=sns.color_palette()[0]
sns.boxplot(data=df,x="host_identity_verified",y="review rate number",color=base_color);
plt.xlabel('Host Verification Status');
plt.ylabel('Review Rate Number');
plt.title('Average Review Rate for each Verification Statistics.');
```
